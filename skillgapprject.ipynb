```python
import pandas as pd

# Load your dataset
df = pd.read_csv("skillgap_survey_sample.csv")
```

This helps you see column names, data types, and any missing values.(Inspect the dataset)

```python
# Check basic info
df.info()


```

```python
# Preview first 5 rows
df.head()
```

```python
# Check shape (rows, columns)
df.shape
```

Columns with high missing percentages might need dropping or imputing (filling).Check for missing values

```python
# Count missing values per column
df.isnull().sum()
```

```python
# Or visualize missing data percentage
(df.isnull().sum() / len(df)) * 100
```

Ensures each participant’s data appears once.(Check for duplicates)

```python
# Count duplicates
df.duplicated().sum()
```

```python
# Remove duplicates
df = df.drop_duplicates(subset=['Respondent_ID'])
```

Check for inconsistent text values

Example for Yes/No, Gender, etc.

```python
# Check unique values in key columns
df['Gender'].unique()
df['Completed_certifications'].unique()
```

```python
# Standardize values
df['Gender'] = df['Gender'].str.strip().str.capitalize()
df['Completed_certifications'] = df['Completed_certifications'].replace({
    'yes': 'Yes', 'no': 'No', 'y': 'Yes', 'n': 'No'
})
```

Handle missing values{Fill with mode/mean/median

Drop incomplete rows}

```python
# Fill missing State with mode
df['State'] = df['State'].fillna(df['State'].mode()[0])

# Drop rows missing Employment_status
df = df.dropna(subset=['Employment_status'])
```

Make sure all missing/duplicate/inconsistent issues are fixed.{Verify final data}

```python
df.info()
df.isnull().sum()
df.head()
```

Save cleaned data

```python
df.to_excel("skillgap_cleaned.xlsx", index=False)
```
