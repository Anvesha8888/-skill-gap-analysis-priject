Gives you mean, median, and distribution of numeric columns, and frequency counts for text columns.(Understand the overall data)

```python
df.describe(include='all')
```

Analyze demographics(Gender distribution:)

```python
df['Gender'].value_counts().plot(kind='bar', title='Gender Distribution')
```

Education level:

```python
df['Highest_qualification'].value_counts().plot(kind='barh', title='Education Levels')
```

State-wise responses:

```python
df['State'].value_counts().head(10).plot(kind='bar', title='Top 10 States of Respondents')
```

Employment insights(a) Employment status count:

```python
df['Employment_status'].value_counts().plot(kind='pie', autopct='%1.1f%%', title='Employment Status')
```

(b) Average time to get a job:

```python
df['Months_to_first_job'].mean()
```

This gives you how long, on average, it takes graduates to find jobs.

Skill gap analysis
a) Compare skill scores

```python
skills = ['Technical_score', 'Problem_solving_score', 'Communication_score', 'Teamwork_score', 'Interview_prep_score']
df[skills].mean().plot(kind='bar', title='Average Skill Scores of Youth')
```

(b) Correlation check

```python
df[skills + ['College_preparedness_score']].corr()
```

Helps you see which skills are most linked to job success or confidence.

Main reasons for not getting jobs

```python
df['Main_reason_for_not_getting_job'].value_counts().plot(kind='barh', title='Top Reasons for Job Struggle')
```

This is one of your key insights â€” the root causes of employability issues.
